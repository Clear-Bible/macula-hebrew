```python
from lxml import etree
import os
lowfat_trees = []
lowfat_ids = []
lowfat_files = os.walk('../lowfat/')
for root, dirs, lowfat_files in lowfat_files:
    for name in lowfat_files:
        if name.endswith('.xml') and not(name.startswith('macula-hebrew')) and not(name.startswith('.')):
            uri = os.path.join(root, name)
            lowfat_tree = etree.parse(uri)
            lowfat_trees.append(lowfat_tree)
            for w_id in lowfat_tree.xpath('//w/@xml:id'):
                lowfat_ids.append(w_id)

# #nodes_tree = etree.parse('../nodes/macula-hebrew.xml')
# with open('../nodes/macula-hebrew.xml') as nodes_file: 
#     nodes_tree = etree.parse(nodes_file)
nodes_trees = []
nodes_ids = []
nodes_files = os.walk('../nodes/')
for root, dirs, lowfat_files in nodes_files:
    for name in lowfat_files:
        if name.endswith('.xml') and not(name.startswith('.')) and not(name.startswith('macula-hebrew')):
            uri = os.path.join(root, name)
            nodes_tree = etree.parse(uri)
            nodes_trees.append(nodes_tree)
            for m_id in nodes_tree.xpath('//m/@xml:id'):
                nodes_ids.append(m_id)

print('Number of lowfat files:', '\t', len(lowfat_trees))
print('Number of nodes files:', '\t', len(nodes_trees))
```

```python
print('Number of lowfat ids:', '\t', len(lowfat_ids))
print('Number of nodes ids:', '\t', len(nodes_ids))

# which nodes ids are not in lowfat
lowfat_missing_ids = set(nodes_ids) - set(lowfat_ids)
# for node_id in nodes_ids:
#     if node_id not in lowfat_ids:
#         print(node_id)
print('Number of missing ids:', '\t', len(lowfat_missing_ids))
print('Missing ids:')
for id in sorted(lowfat_missing_ids):
    print(id)
```

```python
# for each missing id, find the m by searching the nodes_trees, and return the parent::Node of the m with the missing id
for id in lowfat_missing_ids:
    path = f"//m[@xml:id='{id}']/parent::Node"
    print(id)
    for tree in nodes_trees:
        results = tree.xpath(path)
        if len(results) > 0:
            print(etree.tostring(results[0]))
            break
```
